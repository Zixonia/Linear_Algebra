# The Homogenous Systems
---
- In previous discussion we have tackled systems of linear equations that can have either one of the following:
    1. A unique solution
    2. Infinitely many solutions
    3. No solutions or inconsistent.
    
- There are systems of linear equations that have atleast one solutions, and these are called homogenous systems.
- In a homogenous system, the constant term is zero.
- Therefore the augemented matrix of a homogenous system is $[\textbf{A}|0]$.
#### Example
---
$
\begin{equation}
2x+3y-z= 0 \\
-x+5y+2z = 0 \\
\end{equation}$


```python
# Pythonic Solutions
# the following functions are the steps involved in forward elimination
import numpy as np

def RowSwap(A,k,l):
# =============================================================================
#     A is a NumPy array.  RowSwap will return duplicate array with rows
#     k and l swapped.
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A
    
    B = np.copy(A).astype('float64')
        
    for j in range(n):
        temp = B[k][j]
        B[k][j] = B[l][j]
        B[l][j] = temp
        
    return B

def RowScale(A,k,scale):
# =============================================================================
#     A is a NumPy array.  RowScale will return duplicate array with the
#     entries of row k multiplied by scale.
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A
    
    B = np.copy(A).astype('float64')

    for j in range(n):
        B[k][j] *= scale
        
    return B

def RowAdd(A,k,l,scale):
# =============================================================================
#     A is a numpy array.  RowAdd will return duplicate array with row
#     l modifed.  The new values will be the old values of row l added to 
#     the values of row k, multiplied by scale.
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A
    
    B = np.copy(A).astype('float64')
        
    for j in range(n):
        B[l][j] += B[k][j]*scale
        
    return B
```

- To complete the process of determining the solution, create a function or steps of function that will automate backward substitution.

```python
import numpy as np
A = np.array([[1,-1,1,3],[0,1,2,4],[0,0,1,2],[0,0,0,0],[0,0,0,0]]) 

def backwardsub(A):
    m, n = A.shape  

    while m > 0 and np.all(A[m-1, n-2:] == 0):
        m -= 1

    if m == 0:
        return "No Solution..."
    elif n - 1 > m:
        return "Tnfinite Solutions..."

    sol = np.zeros(m)

    for x in range(m - 1, -1, -1):
        coefficients = A[x, :-1]
        constant = A[x, -1]

        if np.all(coefficients == 0):
            return "No Unique Solution..."

        r = np.dot(sol, coefficients)
        sol[x] = (constant - r) / coefficients[x]

    return sol

result = backwardsub(A)
print(result)
```

# Spanning Sets
---
- Recall that the linear equation has the form:
$a_1x_1+a_2x_2+...+a_nx_n=b$
- If we let $a_1,a_2,...,a_n$ be vectors, and $b$ is also a vector of the same size as the $a_1,a_2,...,a_n$.
- If we denotes $\{a_1,a_2,...,a_n\}$ as a set of vectors $\in \mathbb{R}^n$, then any linear combination of $\{a_1,a_2,...,a_n\}$ is called a **span** of $x_1,x_2,...,x_n$, which is denoted by $span(a_1,a_2,...,a_n)$
- If $span(a_1,a_2,...,a_n) \in \mathbb{R}^n$, then it is called the *spanning set* for $\mathbb{R}^n$.
#### Example:
Show that
\begin{pmatrix}
    \begin{bmatrix} 2 \\ -1 \end{bmatrix} \ 
    \begin{bmatrix} 1 \\ 3 \end{bmatrix}\\
\end{pmatrix}
is a span.<br>
Solution:
1. Write the span in terms of a linear equation:
\begin{equation}
    \textbf{x}\begin{bmatrix} 2 \\ -1 \end{bmatrix} \ +
    \textbf{y}\begin{bmatrix} 1 \\ 3 \end{bmatrix} \ =
    \begin{bmatrix} a \\ b \end{bmatrix} \\
\end{equation}
2. Transform the linear equation into its reduved echelon form:
$
\begin{pmatrix}
    2 & 1 &\bigm| & a \\
    -1 & 3 &\bigm| & b \\
\end{pmatrix}
\xrightarrow{\text{R}_{1} \leftrightarrow \text{R}_{2}}
\begin{pmatrix}
    -1 & 3 &\bigm| & b \\
    2 & 1 &\bigm| & a \\
\end{pmatrix}
$
<br><br>
$
\begin{pmatrix}
    -1 & 3 &\bigm| & b \\
    2 & 1 &\bigm| & a \\
\end{pmatrix}
\xrightarrow{\text{R}_{2} 2+ \text{R}_{1}}
\begin{pmatrix}
    -1 & 3 &\bigm| & b \\
    0 & 7 &\bigm| & a+2b \\
\end{pmatrix}
$
<br><br>
$
\begin{pmatrix}
    -1 & 3 &\bigm| & b \\
    0 & 7 &\bigm| & a+2b \\
\end{pmatrix}
\xrightarrow{\frac{1}{7} \text{R}_{2}}
\begin{pmatrix}
    -1 & 3 &\bigm| & b \\
    0 & 1 &\bigm| & \frac{a+2b}{7} \\
\end{pmatrix}
$
<br><br>
$
\begin{pmatrix}
    -1 & 3 &\bigm| & b \\
    0 & 1 &\bigm| & \frac{a+2b}{7} \\
\end{pmatrix}
\xrightarrow{\text{R}_{1}-3 \text{R}_{2}}
\begin{pmatrix}
    -1 & 0 &\bigm| & \frac{b+3a}{7} \\
    0 & 1 &\bigm| & \frac{a+2b}{7} \\
\end{pmatrix}
$
<br><br>
$
\begin{pmatrix}
    -1 & 0 &\bigm| & \frac{b+3a}{7} \\
    0 & 1 &\bigm| & \frac{a+2b}{7} \\
\end{pmatrix}
\xrightarrow{(-1)\text{R}_{1}}
\begin{pmatrix}
    1 & 0 &\bigm| & -\frac{b+3a}{7} \\
    0 & 1 &\bigm| & \frac{a+2b}{7} \\
\end{pmatrix}
$
<br><br>
- The final reduced row echelon is of the form:
$
\begin{pmatrix}
    -1 & 0 &\bigm| & \frac{b+3a}{7} \\
    0 & 1 &\bigm| & \frac{a+2b}{7} \\
\end{pmatrix}
\xrightarrow{(-1)\text{R}_{1}}
\begin{pmatrix}
    1 & 0 &\bigm| & -\frac{b+3a}{7} \\

$\therefore$ it is a span
